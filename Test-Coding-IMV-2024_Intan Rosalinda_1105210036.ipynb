1. Lakukan Preprocessing berupa Flip Horizontal, Flip Vertikal, dan Transpose pada gambar emma.jpg lalu simpan masing-masing hasil Preprocessing tersebut dalam file baru yaitu emma_flip_vertikal.jpg, emma_flip_vertikal.jpg, emma_transpose.jpg. [10 POINTS]

```python
from numpy import expand_dims
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot
```

```python
image = load_img('/content/drive/MyDrive/emma.jpg')
```

```python
dataImage = img_to_array(image)
```

```python
imageNew = expand_dims(dataImage, 0)
```

```python
imageDataGen = ImageDataGenerator(horizontal_flip=True)
```

```python
iterator = imageDataGen.flow(imageNew, batch_size=1)
```

```python
for i in range(6):

	pyplot.subplot(330 + 1 + i)
	batch = iterator.next()
	image = batch[0].astype('uint8')
	pyplot.imshow(image)

pyplot.show()
```

```python
imageDataGen = ImageDataGenerator(vertical_flip=True)
```

```python
iterator = imageDataGen.flow(imageNew, batch_size=1)
```
